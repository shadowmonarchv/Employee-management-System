```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.set_title("Basic Figure and Axis")
ax.set_xlabel("Days")
ax.set_ylabel("Total Patients")
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
recovery_rate = [65, 70, 60, 78, 80, 85]

plt.bar(months, recovery_rate, color=['red', 'blue', 'green', 'purple', 'orange'])

plt.xlabel('Months')
plt.ylabel('Recovery Percentage')
plt.title("Patient Recovery Trends")
plt.legend()
plt.show()
```

```python
ages = np.random.randint(10,80,100)

plt.hist(ages, bins=10, color='blue', edgecolor='black')
plt.xlabel('Age Groups')
plt.ylabel('Number of Patients')
plt.title("Age Distribution of Patients")
plt.show()
```

```python
from wordcloud import WordCloud
symptoms_text= "Fever cough cold headache nausea fatigue pain weakness dizziness allergy infection sore throat congestion"
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(symptoms_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Common Symptoms Analysis")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("telecom_customer_data.csv")
df.head()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x="Call_Duration", y="Satisfaction_Score", hue="Network_Provider")

plt.legend(title="Network Provider", bbox_to_anchor=(1, 1), loc="upper left")

plt.xlabel("Call Duration (minutes)")
plt.ylabel("Satisfaction Score")
plt.title("Call Duration vs Satisfaction Score")

plt.show()
```
