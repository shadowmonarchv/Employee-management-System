### ðŸ§ª Hypothesis Test 1: One-Sample, One-Tailed t-test

ðŸ“Œ Scenario:

A company claims that the average delivery time is less than 30 minutes. 

You collect sample delivery times to verify the claim.

```python
from scipy import stats
import numpy as np

# Sample delivery times
delivery_times = np.array([28, 27, 32, 30, 29, 26, 31, 28, 27, 29])

# One-sample t-test (H0: mean = 30)
t_stat, p_val = stats.ttest_1samp(delivery_times, popmean=30)

# One-tailed p-value for mean < 30
one_tailed_p = p_val / 2
alpha = 0.05

print(f"Test statistic: {t_stat:.4f}")
print(f"One-tailed p-value: {one_tailed_p:.4f}")

```

```python
# what is your inference
```

```python
if one_tailed_p < alpha:
    print("âœ… Reject H0: Delivery time is significantly less than 30 minutes.")
else:
    print("âŒ Fail to reject H0: Not enough evidence to support itâ€™s less than 30.")
```

### ðŸ§ª Hypothesis Test 2: One-Sample, Two-Tailed t-test

ðŸ“Œ Scenario:

A factory claims that the mean weight of packets is 100 grams. 

You want to check if it is different (either less or more) from the claim.

```python
weights = np.array([102, 98, 101, 97, 103, 99, 100, 96, 104, 98])
t_stat, p_val = stats.ttest_1samp(weights, popmean=100)
alpha = 0.05

print(f"Test statistic: {t_stat:.4f}")
print(f"Two-tailed p-value: {p_val:.4f}")

```

```python
# your inference
```

```python
if p_val < alpha:
    print("âœ… Reject H0: Mean packet weight is significantly different from 100g.")
else:
    print("âŒ Fail to reject H0: No evidence of a significant difference.")
```

### ðŸ§ª Hypothesis Test 3: Two-Sample, Two-Tailed t-test

ðŸ“Œ Scenario:

You want to compare salaries of Data Analysts and Software Engineers to see if there is a significant difference.


```python
analyst_salaries = np.array([48000, 50000, 52000, 51000, 49500])
engineer_salaries = np.array([60000, 61000, 59000, 62000, 60500])

t_stat, p_val = stats.ttest_ind(analyst_salaries, engineer_salaries, equal_var=False)
alpha = 0.05

print(f"Test statistic: {t_stat:.4f}")
print(f"Two-tailed p-value: {p_val:.4f}")

```

```python
#inference
```

```python
if p_val < alpha:
    print("âœ… Reject H0: There is a significant difference in average salaries.")
else:
    print("âŒ Fail to reject H0: No significant difference in average salaries.")
```

### ðŸ§ª Hypothesis Test 4: Two-Sample, One-Tailed t-test
ðŸ“Œ Scenario:

A new teaching method claims students score higher than students taught with the old method.

You want to test if new method scores > old method.


```python
from scipy import stats
import numpy as np

# Sample delivery times
delivery_times = np.array([28, 27, 32, 30, 29, 26, 31, 28, 27, 29])

# One-sample t-test (H0: mean = 30)
t_stat, p_val = stats.ttest_1samp(delivery_times, popmean=30)

# One-tailed p-value for mean < 30
one_tailed_p = p_val / 2
alpha = 0.05

print(f"Test statistic: {t_stat:.4f}")
print(f"One-tailed p-value: {one_tailed_p:.4f}")
```

```python
#inference
```

```python

if (t_stat < 0) and (one_tailed_p < alpha):
    print("âœ… Reject H0: Delivery time is significantly less than 30 minutes.")
else:
    print("âŒ Fail to reject H0: Not enough evidence to support itâ€™s less than 30.")
```
