```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

```

```python
text = "This is not a good product, but it is affordable."
tokens = word_tokenize(text) # Splits the text into individual words/punctuation.

stop_words = set(stopwords.words('english')) #  Loads English stopwords (common words like "is", "a") as a set.

filtered = [word for word in tokens if word.lower() not in stop_words]
# Removes stopwords from tokens, keeping only meaningful words.

print("Without Stop Words:", filtered)
```
