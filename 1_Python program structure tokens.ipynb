##  Hands-on Problem solving: Retail Sales Calculation

### calculate total price of tshirt after discount

```python
# tshirt_discount = 10 % 
tshirt_price = 1500
tshirt_discount = 10 
final_price = tshirt_price - (tshirt_discount/100)*tshirt_price

print("T shirt Price:", tshirt_price)
print("Discount Applied:", tshirt_discount)
print("Final Price after Discount:", final_price)

```
