### Recursive function

###  Recursive Function:

### A function that calls itself to solve a smaller part of the problem.

###  Usage:

### Useful for problems that can be broken into smaller, similar sub-problems.
### Common examples: factorial, Fibonacci, tree traversal, etc.

###  Limitation:

### - May lead to stack overflow (maximum recursion depth exceeded) for large inputs.
### - Slower than loops in some cases unless optimized with memoization.


### Write a program to calculate factorial of given number using Recursive function

```python
def myfactorial(n):
    # Base case: factorial of 0 is 1
    if n == 1:
        return 1
    # Recursive case: factorial of n is n multiplied by factorial of n-1
    else:
        return n * myfactorial(n - 1)
```

```python

num = int(input("Enter a number to find its factorial: "))
result = myfactorial(num)
print("Factorial of", num, "is", result)
```

### Write a program in python with recursive function , to add all numbers between 1 and n

5=>

1+2+3+4+5=15 

5+4+3+2+1+0 =15

```python
def sum_of_natural_numbers(n):
    
    if n == 1:  # Base case: sum of first natural number is the number itself
        return 1
    else:     # Recursive case: sum of first n natural numbers is n plus sum of first n-1 natural numbers
        return n + sum_of_natural_numbers(n - 1)
```

```python
# Example usage:
num = int(input("Enter a number to find the sum of its natural numbers: "))
result = sum_of_natural_numbers(num)
print("Sum of first", num, "natural numbers is", result)
```

###  Lambda and Map Functions in Python 



###  1. What is Lambda Function?

### A lambda function is an anonymous (unnamed) function defined in a single line using the keyword 'lambda'.

### It is mainly used for small, simple functions that are used temporarily.



```python
#  Syntax:
# lambda arguments: expression

#  Example: Add 10 to a number using lambda

add10 = lambda x: x + 10

print("Result using lambda:", add10(5))  # Output: 15
```

```python

#  2. Same function using def (without lambda)
def add10_def(x):
    return x + 10

print("Result using def:", add10_def(5))  # Output: 15
```


###  When to use lambda:
### - For quick, throwaway functions
### - When using functions like map(), filter(), sorted(), etc.

###  When NOT to use lambda:
### - For complex logic or multiple lines of code



###  3. What is map() function?
### The map() function applies a function to each item in an iterable (like a list).

###  Syntax:
### map(function, iterable)

###  Example: Double each number in a list


```python
numbers = [1, 2, 3, 4]

def twice(x):
    return x * 2

result = map(twice, numbers)
print("Doubled numbers (using def):", list(result))

```

###  4. Using lambda with map()

```python
# This avoids defining a separate function
result_lambda = map(lambda x: x * 2, numbers)
print("Doubled numbers (using lambda):", list(result_lambda))
```

###  Another Example: Convert list of strings to uppercase

```python
words = ['apple', 'banana', 'mango']
uppercase_words = map(lambda w: w.upper(), words)
print("Uppercase words:", list(uppercase_words))
```

###  5. Real-world mini case: Adding 18% GST to item prices

```python
item_prices = [100, 250, 80, 120]

# Add 18% GST using lambda with map
final_prices = map(lambda p: round(p * 1.18, 2), item_prices)
print("Prices after GST:", list(final_prices))
```
