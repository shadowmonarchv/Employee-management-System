# Time Series Forecasting : - Naïve Forecasting (Simplest Forecasting method)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Sales": [200, 220, 240, None]  # April is unknown
}
df = pd.DataFrame(data)
df
```

```python

# Plot existing data
plt.plot(df["Month"], df["Sales"], marker='o', label="Actual Sales")

plt.title("Naïve Forecast Example")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)

# Naïve forecast
naive_forecast = df["Sales"].iloc[2]  # Last known value (March)

plt.plot("Apr", naive_forecast, marker='x', markersize=10, color='red', label="Forecast (Naïve)")

plt.legend()
plt.show()

print(f"Naïve Forecast for April: {naive_forecast}")
```
