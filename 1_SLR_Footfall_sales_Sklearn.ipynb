# Write a machine learing algorithm to predict daily salses based on footfalls in General store


```python
# Import required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression 

# sklearn A popular Python library for machine learning, built on NumPy, SciPy, and matplotlib.
# linear_model A submodule within sklearn focused on linear modeling techniques.
# LinearRegression (class) What it is: A supervised learning algorithm for predicting numeric values (regression)
import matplotlib.pyplot as plt

```

```python
# Load the dataset
data = pd.read_csv('C:\\Users\\askpr\\Tutedude\\ML\\SLRegression\\General_Stores.csv')

# Preview the data
print("Dataset preview:")
data.head(3)
```

```python
# Define independent (Footfalls) and dependent (DailySales) variables
```

```python

X = data[['Footfalls']]
Y = data['DailySales']
```

```python
X
```

```python
Y
```

X (Features/Inputs)

Double brackets ([['Footfalls']]) are used to select a DataFrame with two dimensions (rows and columns).

Scikit-learn expects the input features (X) to be in a 2D array-like structure, even if there is only one feature (column).

Y (Target/Output)

Single brackets (['DailySales']) are used to select a Series, which is a 1D structure (just one column).

The target variable (Y) should be a 1D array or Series.

```python
#%matplotlib inline
plt.xlabel('Footfalls')
plt.ylabel('Daily Sales')
plt.scatter(X,Y,color='red',marker='o')
```

```python
`# Create and train the linear regression model

model = LinearRegression()
model.fit(X, Y)
```

```python
# fit is a method for creating model or trainning my data or creating relations between x and y
```

```python
# Print the model parameters
print(f"Coefficient (slope): {model.coef_}")
print(f"Intercept: {model.intercept_}")

```

### Predict sales  using the trained model

```python
# Prepare new footfall values (for example: 45, 50, 60)
new_footfalls = pd.DataFrame({'Footfalls': [45, 50, 60]})
new_footfalls
```

```python

# Predict new sales
predicted_sales = model.predict(new_footfalls)

print("Predicted Daily Sales for new Footfalls:\n\n")

print(pd.DataFrame({'Footfalls': new_footfalls['Footfalls'], 'PredictedSales': predicted_sales}))
```
