```python
import nltk
nltk.download('punkt')  # Sentence and word tokenizers

#punkt is a pretrained model in NLTK used for tokenizing text into sentences and words.

# You are downloading this trained tokenizer model to your local machine.
# It‚Äôs not part of the default NLTK install.

from nltk.tokenize import word_tokenize, sent_tokenize
```

```python

# üéØ Input text
text = "Hello! I'm learning NLP. Let's test tokenization ‚Äî sentence-wise and word-wise. Amazing, isn't it?"

print("=== üìå NLTK Sentence Tokenization ===")

sentences = sent_tokenize(text)

# sent_tokenize: Splits text into sentences using NLTK's pre-trained Punkt tokenizer
# - Handles abbreviations, punctuation, and multiple languages
# - Returns list of sentences
```

```python
sentences
```

```python
for i, sent in enumerate(sentences):
    print(f"Sentence {i+1}: {sent}")


# enumerate adds a counter (i) to each element (sent) in the iterable (sentences), 

# allowing you to loop with both index and value.

# (Example: Turns ['Hello', 'World'] into (0, 'Hello'), (1, 'World').)

    
print("\n=== üîç NLTK Word Tokenization ===")

for i, sent in enumerate(sentences):
    print(f"\nSentence {i+1} tokens:")
    words = word_tokenize(sent)
    print(words)
    
    
    
# word_tokenize: Splits text/sentences into words (tokens)
    # - Handles contractions ("I'm" ‚Üí "I", "'m")
    # - Splits punctuation into separate tokens
    # - Uses TreebankWordTokenizer by default
```

```python
#===============================
# üöÄ spaCy Tokenization Example
# ===============================
import spacy
nlp = spacy.load("en_core_web_sm")  # Load small English model

doc = nlp(text)

print("\n=== üöÄ spaCy Tokenization ===")
print("Tokens:")
print()
print()
for token in doc:
    print(token.text)
print()
print()
print("\n=== ‚úÇÔ∏è spaCy Sentence Segmentation ===")

for i, sent in enumerate(doc.sents):
    print(f"Sentence {i+1}: {sent.text}")
```
