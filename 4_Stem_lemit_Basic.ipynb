```python
from nltk.stem import PorterStemmer

#nltk → The Natural Language Toolkit library for text processing.

#stem → The NLTK submodule containing stemming algorithms.

# PorterStemmer → A class implementing the Porter stemming algorithm (reduces words to root forms, e.g., "running" → "run").
```

```python

stemmer = PorterStemmer()
print(stemmer.stem("running"))         # run
print(stemmer.stem("responsibilities")) # respons
print(stemmer.stem("flies"))            # fli


#Stems "running" → "run" (base form).
```

```python
from nltk.stem import WordNetLemmatizer

#nltk → The Natural Language Toolkit library for text processing.

#stem → The NLTK submodule containing stemming algorithms.

# WordNetLemmatizer ->A class implementing the lemmat algorithm (reduces words to root forms, e.g., "glies" → "fly").
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')

```

```python
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("running", pos="v"))  # run
print(lemmatizer.lemmatize("better", pos="a"))   # good
print(lemmatizer.lemmatize("flies",pos='v'))   # fly

# pos="v" → Specifies the part of speech (verb here) for accurate lemmatization (default is noun, "n").
```

```python
# Full example: Stopword removal + Lemmatization
from nltk.tokenize import word_tokenize

from nltk.corpus import stopwords

# Download if not already
nltk.download('stopwords')

# Get English stop words
stop_words = stopwords.words('english')

text = "Cats are running around the house, chasing mice happily."

tokens = word_tokenize(text)
filtered = [word for word in tokens if word.lower() not in stop_words]

lemmatized = [lemmatizer.lemmatize(word.lower(), pos='v') for word in filtered]
print("Processed:", lemmatized)
```
