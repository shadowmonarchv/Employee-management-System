## üîÑ `while` Loop Basics

### üîç What is a `while` loop?
A `while` loop keeps executing the block of code as long as the condition is `True`.

**Syntax:**
```python
# initiasiation
while condition:
    # code block
    # updation
```


### ‚úÖ Use-Cases for `while` loop
- When we **don‚Äôt know how many times** we need to run the loop.
- Used in **input-driven** or **sensor-based** applications.
- Useful for **menu-driven** or **event-based** programs.


````python
# Example: Count from 1 to 5 using while loop
i = 1 #initialisation
while i <= 5: # condition
    print(i)
    i += 1 # updation
````

### üéì Basic Program: Sum of Numbers from 1 to n

````python
n = int(input("Enter a number: "))
i = 1
total = 0
````

````python
while i <= n:
    total += i
    i += 1
````

````python
print("Sum of numbers from 1 to", n, "is", total)
````

## üîÅ `for` Loop Basics

### üîç What is a `for` loop?
Used when we want to **repeat a block of code a fixed number of times**.

**Syntax:**
```python
for variable in sequence:
    # code block
```


### üîß Using `range()` with `for`
- `range(n)` ‚Üí 0 to n-1  
- `range(start, stop)` ‚Üí from start to stop-1  
- `range(start, stop, step)` ‚Üí step can be positive or negative


````python
for i in range(1,5,2):
    print(i)
````

````python
# Example: Print numbers from 1 to 5 using for loop
for i in range(1, 6):
    print(i)
````
