## 1) `if` Statement
**Concept:** Executes code only if the condition is `True`.
** Example:** Checking if a person is eligible to vote (Age  18).

```python
a=119
if a>20:
    print('greater')
```

```python
a=10
if a==100:
    print('equal')
    print('good bye')
else:
    print('not equal')
```

```python
age = 20
if age >= 18:
    print("You can vote in  elections! ")
```

## 2) `if-else` Statement
**Concept:** Executes one block if `True`, else another block.
** Example:** Checking if a movie ticket is priced as "Adult" or "Child".

```python
age = 15
if age >= 18:
    print("Ticket price: 200 (Adult)")
else:
    print("Ticket price: 100 (Child)")
```

## 3) `elif` Statement
**Concept:** Checks multiple conditions in sequence.
** Example:** Grading students based on marks (90+ = A, 80+ = B, etc.).

```python
marks = int(input('enter marks')) 
```

```python
if marks >= 90:
    print("Grade: A+ (Topper!)")
elif marks >= 80:
    print("Grade: B (Good)")
elif marks >= 60:
    print("Grade: C (Average)")
else:
    print("Grade: D (Improve)")
```

## Business Case Study: Flipkart Discount Strategy
**Problem:** Flipkart wants to give discounts based on purchase amount:
- 5000+  20% off
- 3000+  10% off
- Else  No discount

**Solution:**

```python
purchase_amount = 4500
if purchase_amount >= 5000:
    discount = 0.2 * purchase_amount
    print(f"Discount: {discount} (20% off)")
elif purchase_amount >= 3000:
    discount = 0.1 * purchase_amount
    print(f"Discount: {discount} (10% off)")
else:
    print("No discount. Shop more!")

final_price = purchase_amount - discount
print(f"Final price: {final_price}")
```
